```python
#Import Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("Hiv_positive.csv")
df1
```

```python
df2 = pd.read_csv("Hiv_Death.csv")
df2
```

```python
df3 = pd.read_csv("Tb_Hiv_positive.csv")
df3
```

```python
df4 = pd.read_csv("Tb_Hiv_Death.csv")
df4
```

#  cleaning and merging 

```python
df3 = df3.drop(columns={"Unnamed: 0"})
df3
```

```python
df3 = df3.rename(columns={'Entity':'Country','TB patients tested positive for HIV - Total':'Tb-Hiv'})
df3
```

```python
df4 = df4.drop(columns={"Unnamed: 0"})
df4
```

```python
df4 = df4.rename(columns={'Entity':'Country','Estimated TB-related deaths among people living with HIV - estimate':'Tb-Hiv-Death'})
df4
```

```python
df1 = df1.drop(columns={"Unnamed: 0","Count"})
df1
```

```python
df2 = df2.drop(columns={"Unnamed: 0","Count"})
df2 
```

```python
df1 = df1.rename(columns={"Count_median":"aliveCount"})
df1
```

```python
df2.rename(columns={"Count_median":"Death","Count_min":"Death_min","Count_max":"Death_max"},inplace=True)
df2
```

```python
#Merging first two data frames
df5 = pd.merge(df1,df2, how="left", on=["Country", "Year","WHO Region"])
df5
```

```python
#Merging next two data frames
df6 =  pd.merge(df3,df4, how="left", on=["Country","Code","Year"])
df6
```

```python
df = pd.merge(df5,df6, how="left", on=["Country","Year"])
df
```

```python
#Data types in data frame
df.dtypes
```

```python
#Convert year data type to integer
df['Year'] = df['Year'].astype(int)
df
```

```python
# Creating new column 
df['Non-Tb'] = df['Death']-df['Tb-Hiv-Death']
df
```

```python
#Naming the index
df.index.name = 'Index'
df
```

```python
#Drop columns
df = df.drop(columns={"Count_min","Count_max","Death_min","Death_max"})
df
```

# Visualization

univariate analysis

numerical variables

```python
def plottingHistogram(vari):
    plt.figure()
    plt.hist(df[vari])
    plt.xlabel(vari)
    plt.title(f"data_frequency - {vari}")
    plt.ylabel("freq")
    plt.show    
```

```python
columns = ['aliveCount','Death','Tb-Hiv','Year','Tb-Hiv-Death']
for c in range(len(columns)):
    plottingHistogram(columns[c])
```

## for categorical variables

```python
def plotBar(v):
    count=2
    var_Value = df[v].value_counts()
    if v=='Year':
        var_Value = var_Value[:2]
    else:
        var_Value = var_Value[:5]
    
    #visualize
    plt.figure()
    plt.bar(var_Value.index,var_Value)
    plt.xticks(rotation=45)
    plt.ylabel('Fre')
    plt.title(f'{v}')
    plt.show()
col = ['WHO Region']
for va in range(len(col)):
    plotBar(col[va])
    
    
    
```
