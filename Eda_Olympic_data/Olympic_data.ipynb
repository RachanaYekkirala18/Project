```python
df = df.drop(["ID","Games"], axis=1)
df
```

# Data Filling

```python
df['Medal'] = df['Medal'].fillna(value='No Medal')
```

```python
avg = np.round(np.mean(df['Age']),1)
df['Age'] = df['Age'].fillna(avg)
```

```python
import numpy as np
imp_col = ['Height','Weight']
for col in imp_col:
    df[col] = df[col].fillna(np.mean(df[col]))
    df[col] = np.round(df[col],1)
# df['Weight']=df['Weight'].fillna(df['Weight'].m().iloc[0])
```

```python
def plotHistogram(variable):
    plt.figure()
    plt.hist(df[variable],bins=85,color="purple")
    plt.xlabel(variable.upper())
    plt.ylabel("Frequency")
    plt.title(f"Data Frequency - {variable}")
    plt.show    
```

```python
category = ['Age','Height','Weight','Year']
for c in category:
    plotHistogram(c)
```

```python
category = ['Age','Height','Weight','Year']
def plotBoxGraph(veriable):
  
    plt.boxplot(df[veriable])
    plt.title(f"Box plot for {veriable} veriable")
    plt.xlabel(veriable)
    plt.ylabel('Value')
    plt.show()

for c in category:

    plotBoxGraph(c)
```

# categorical variables

```python
def plotBar(veriable,n=5):
  
    varValue = df[veriable].value_counts()
    varValue = varValue[:5]
    
    #visualize
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index,varValue,color="pink")
    plt.xticks(rotation=45)
    plt.ylabel('Frequency')
    plt.title(veriable)
    plt.show()
    
   
```

```python
categorical = ['Sex','Team','City']
for c in categorical:
    plotBar(c)
```
